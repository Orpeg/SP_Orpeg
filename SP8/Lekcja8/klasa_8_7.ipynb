```python
# break 
# continue

lst = [12,20,29,-20,15,14,9,35]
# liczymy sumę licz w liście
sum = 0
for e in lst:
    sum+=e
print('suma liczb w liście wynosi sum = ', sum)
```

```python
sum = 0
for e in lst:
    if e<0:
        break
    sum+=e
print('suma pierwszych dodatnich wyrazów wynosi sum = ', sum)
```

```python
# continue
sum = 0
for e in lst:
    if e<0:
        continue
    sum+=e
print('suma wyrazow tylko dodatnich sum =', sum)
```

```python
# silnia
# 1*2*3...*N
# METODA ITERACYJNA
s = 1
n = int(input('Podaj liczbe  n = '))
for i in range(1,n+1):
    s *= i
print('silnia = ', s)
```

```python
# silnia(n)= silnia(n-1)*n
# metoda rekurencyjna

k = int(input('Podaj liczbę n = '))
def silnia(n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    return silnia(n-1)*n
print('silnia metoda rekurencyja = ', silnia(k))
```
